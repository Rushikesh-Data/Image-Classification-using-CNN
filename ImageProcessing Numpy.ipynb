```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
image_path = "C:/Users/HP/Bgkd.jpg"
image=plt.imread(image_path)
```

```python
image_array=np.array(image)
```

```python
print("shape of numpy array:", image_array.shape)
```

```python
plt.figure(figsize=(30,10))
plt.subplot(2,3,1)
plt.title("Original Image")
plt.imshow(image_array)
plt.axis("off")
```

```python
horizontal_flip = np.flip(image_array, axis=1)
plt.figure(figsize=(30,10))
plt.subplot(2, 3, 3)
plt.title("Horizontal Flip")
plt.imshow(horizontal_flip)
plt.axis("off")
```

```python
def resize_image(image_array, new_width, new_height):
    image = Image.fromarray(image_array)  
    resized_image = image.resize((new_width, new_height), Image.Resampling.LANCZOS)
    return np.array(resized_image)

# New dimensions 
new_width = 100
new_height = 80

resized_image_array = resize_image(image_array, new_width, new_height)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image_array)
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title(f"Resized Image ({new_width}x{new_height})")
plt.imshow(resized_image_array)
plt.axis("off")

plt.tight_layout()
plt.show()
```
